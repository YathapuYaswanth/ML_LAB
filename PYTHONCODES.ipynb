```python
correct_pin = "1234"
attempts = 3

while attempts>0:
  pin = input("enter the pin:")
  if pin == correct_pin:
    print("access granted")
    break
  else:
    attempts -=1
    print(f"incorrect pin number.attempts left:{attempts}")
  if attempts == 0:
    print("card blocked!")
```
